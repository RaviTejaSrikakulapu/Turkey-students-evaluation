## K Means:

```python
X=data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = pd.DataFrame(sc.fit_transform(X), columns=X.columns)
X.head()
```

```python
from sklearn.cluster import KMeans

cluster_range = range(1,16)
cluster_errors = []

for num_clusters in cluster_range:
    clusters = KMeans(num_clusters, n_init=5, max_iter=100)
    clusters.fit(X)
    
    cluster_errors.append(clusters.inertia_)
    
pd.DataFrame({'num_clusters':cluster_range, 'Error': cluster_errors})
```

```python
plt.figure(figsize=(10,5))
plt.plot(cluster_range, cluster_errors, marker = "o" )
plt.title('Elbow Plot')
plt.xlabel('Number of Clusters')
plt.ylabel('Error')
plt.xticks(cluster_range)
plt.show()
```

## PCA:

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(X)
```

```python
X_pca = pca.transform(X)
X_pca.shape
```

```python
pca.components_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(10,6))

plt.plot(range(1,34), np.cumsum(pca.explained_variance_ratio_), color='k', lw=2)

plt.xlabel('Number of components')
plt.ylabel('Total explained variance')

plt.axvline(8, c='b')
plt.axhline(0.9, c='r')

plt.show()
```

```python
pca = PCA(n_components=8)
pca.fit(X)
X_pca = pd.DataFrame(pca.transform(X))
X_pca.shape
```

```python
sns.pairplot(X_pca, diag_kind='kde')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(X_pca.corr(),annot=True,fmt='.1f')
plt.show()
```

```python
from sklearn.cluster import KMeans

cluster_range = range(1,16)
cluster_errors = []

for num_clusters in cluster_range:
    clusters = KMeans(num_clusters, n_init=5, max_iter=100)
    clusters.fit(X_pca)
    
    cluster_errors.append(clusters.inertia_)
    
pd.DataFrame({'num_clusters':cluster_range, 'Error': cluster_errors})
```

```python
plt.figure(figsize=(10,5))
plt.plot(cluster_range, cluster_errors, marker = "o" )
plt.title('Elbow Plot')
plt.xlabel('Number of Clusters')
plt.ylabel('Error')
plt.xticks(cluster_range)
plt.show()
```

```python
X_pca_df = X_pca.copy()
kmeans = KMeans(3, n_init=5, max_iter=100)
kmeans.fit(X_pca_df)
X_pca_df['label'] = kmeans.labels_
X_pca_df['label'].value_counts().sort_index()
```
